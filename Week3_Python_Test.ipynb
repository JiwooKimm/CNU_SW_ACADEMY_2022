```python
def solution(nums):

    def isprimeNum(integer):
        import math
        for j in range(2, int(math.sqrt(integer))+1):
            if integer%j == 0:
                return False
                break
        return True
    
    answer = 0
    
    import itertools as tools
    comb= list(tools.combinations(nums,3))
    ncomb= len(comb)

    
    if ncomb == 1:
        tot= sum(comb[0])
        if isprimeNum(tot) == True:
            answer += 1
            print (list(comb[0]),'  ', tot,'()   . n:', answer)
    else:
        for i in range(ncomb):
            tot= sum(comb[i])
    #         print(tot, comb[i])

            if isprimeNum(tot) == True:
                answer += 1
                print (list(comb[i]),'  ', tot,'()   . n:', answer)            
    return answer

solution(nums)
```
