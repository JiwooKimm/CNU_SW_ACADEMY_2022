### 2. 2차함수 그래프 with `plot()`

Plot data, type as Dataframe, 2D array 

**<span style='color:royalblue'> i. 뚝딱뚝딱 line plot 그려보기 </span>**

```python
x= np.array([1,2,3,4,5])
# y= np.array([1,3,9,16,25])

plt.plot(x,x**2)
```

**<span style='color:royalblue'> ii. 스무스한 line plot 데이터 생성하기 </span>**  

`np.arange(start, end, step)`

```python
# np.arange(a, b, c) c: 0.01 :: 실수 단위 c씩 증가하는 a ~ b 까지의 array 생성
np.arange?
x= np.arange(-10, 10, 0.01)
plt.plot(x,x**2)
```

**<span style='color: royalblue'> iii. x, y축에 label 추가하기 </span>**  

`plt.xlabel()` `plt.ylabel()`

```python
x= np.arange(-10, 10, 0.01)
plt.plot(x,x**2)

##
plt.xlabel('x value')
plt.ylabel('f(x) value')
##

plt.show()
```

**<span style='color:royalblue'> iiii. x,y축의 범위 설정 </span>**  
`plt.axis()`

```python
x= np.arange(-10, 10, 0.01)
plt.plot(x,x**2)

plt.xlabel('x value')
plt.ylabel('f(x) value')

##
plt.axis([-5,5,0,25])
##

plt.show()
```

**<span style='color:royalblue'> v. Subtick 설정 </span>**  
`plt.xticks()` `plt.yticks()`


```python
x= np.arange(-10, 10, 0.01)
plt.plot(x,x**2)

plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5,5,0,25])

##
plt.xticks([i for i in range(-5,5,1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0,27,1)]) # y축의 눈금 설정
## 
plt.show()
```

**<span style='color:royalblue'> vi.Title & Legend 붙이기 </span>**  
`plt.title()`  
`plt.legend()`:plot 그릴 때 먼저 label 지정해줘야함 

```python
x= np.arange(-10, 10, 0.01)
plt.plot(x,x**2, label = 'trend')

plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5,5,0,25])

plt.xticks([i for i in range(-5,5,1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0,27,1)]) # y축의 눈금 설정

##
plt.title('y = x$^2$ graph')
plt.legend()
##

plt.show()
```

## III. Matplotlib Case Study

### 1. Line Plot
`.plot()`
- 연속성 있는 데이터의 가시화에 유용
 e.g., 시계열 데이터
 

```python
x= np.arange(20)
y= np.random.randint(0, 20, 20) # 0 ~ 20까지의 난수 20개
```

```python
plt.plot(x,y)
```

<span style='color:royalblue'> >> 조금 더 가독성 좋게 하면,</span>

```python
plt.plot(x,y)
plt.axis([0,20,0,20])
plt.yticks(np.arange(0,20,5))
```

### 2. Scatter Plot
`.scatter`

- x, y과 완전히 independent한데, dependence를 보고자 할 때 유용

```python
plt.scatter(x,y)
plt.show()
```

### 3. Box Plot
`plt.boxplot()`
- 수치형 데이터에 대한 정보 (e.g., Q1, Q2, Q3, min, max, ...) 가시화에 유용
- 단일 인자, 멀티 인자 모두 가능! >> container 형식으로 묶어서 인자로 전달하면 박스 여러개 그려줌


```python
plt.boxplot((x,y))
plt.title('Boxplot of y')
plt.show()
```

### 4. Bar Plot
`plt.bar()`
- 범주형 데이터의 **값**과 그 **값의 크기**를 직사각형으로 나타냄

```python
plt.figure(figsize=(8,3))
plt.bar(x,y)
plt.xticks(np.arange(0,20,1))
plt.show()
```

### + Histogram
`plt.hist()`
- 도수분포를 직사각형의 막대 형태로 나타내는 점은 barplot과 동일!
- **계급**으로 나타냄: 위 막대처럼 0, 1, 2가 아닌 0~2까지의 <u>범주형</u> 데이터로 구성 후 plot!
- 서로의 범주가 연속적이기 때문에 막대가 이어진 형태!

```python
plt.hist(y, bins= np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

### 5. Pie Chart
`plot.pie()`
- 데이터에서 전체에 대한 **비율** 확인하기 유용!
- 원 & 부채꼴로 비율을 나타냄 
- `labels` option을 지정해서 데이터 개별의 레전드를 표출할 수 있음

```python
z= [100, 300, 200, 400]

plt.pie(z, labels= ['one','two','three','four'], shadow=True)
plt.show()
```

### 0. Seaborn Import 하기

### 1. Kernel Density Plot
`sns.kdeplot()`
- 히스토그램과 같은 연속적인 분포를 곡선화한 plot


```python
x = np.arange(0,22,2)
y= np.random.randint(0,20,20)
```

```python
plt.hist(y)
```

```python
sns.kdeplot(y, shade= True)
plt.show()
```

### 2. Count Plot
`sns.countplot()`
- 범주형 column의 빈도수를 시각화 --> Groupby 후 도수하는 것과 동일!

```python
vote_df= pd.DataFrame({'name':['Andy','Bob','Cat'],'vote':[True, True, False]})
vote_df
```

<span style='color:royalblue'> *먼저 'matplotlib'의 'barplot' 이용해서 그려보자:*  
    *Dictionary type의 'vote_count'에서 True와 False가 몇 개 씩 나왔을가*</span>

```python
# matplotlib barplot
vote_count= vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False, True], height= vote_count['name'])
plt.show()
```

<span style='color:royalblue'>*위와 같이 그린다면 groupby 와 plot을 각각 해줘야함! ~--> 번거롭~<br>
**seaborn**쓰면 원큐에 해결 and bar의 색깔도 자동으로 다르게 설정해줘서 가독성 굿굿임*</span>

```python
sns.countplot(x=vote_df['vote'])
plt.show()
```

### 3. Cat Plot
`sns.catplot(x, y, data)`
- 숫자형 변수나 하나 이상의 범주형 변수의 관계 나타내기에 유용
- *cat* \~ concatration! 데이터들을 묶어서 본다는 것! 
- 해서 복잡한 데이터의 가시화에 좋음

```python
covid= pd.read_csv('/Users/jiwookim/Library/Mobile Documents/com~apple~CloudDocs/Documents/Academy/2022CNU_SW_ACADEMY/Programmers/covid-19/country_wise_latest.csv')
covid.head()
```

```python
s= sns.catplot(x= 'WHO Region', y= 'Confirmed' , data= covid)
s.fig.set_size_inches(10,4)
plt.show()
```

```python
#kind option 지정해주면 violin plot도 할 수 ㅇㅇ!
s= sns.catplot(x= 'WHO Region', y= 'Confirmed' , data= covid, kind= 'violin')
s.fig.set_size_inches(10,4)
plt.show()
```

### 4. Strip Plot
`sns.stripplot()`
- scatterplot 과 유사! >> 데이터의 수치 표현 

```python
sns.stripplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

### + Swarmplot
- 중복값까지 확인 가능!

```python
sns.swarmplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

### 5. Heatmap
`sns.heatmap()`
- 데이터의 행렬를 색상으로 표현해줌
- 가장 많이 쓰일 때는 correlation!
- 'cmap' 옵션 지정해서 색상 표현 가능

```python
plt.figure(figsize=(8,6))
sns.heatmap(covid.corr(), cmap='Greens')
plt.show()
```
